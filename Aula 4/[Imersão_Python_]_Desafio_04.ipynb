```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
dados = yf.download('AAPL', start='2023-01-01', end='2023-12-31')
```

```python
mpf.plot(dados.head(60), type='candle', figsize = (20,12), volume=True, mav=(7,14), style='starsandstripes')
```

```python
mpf.plot(dados.head(60), type='candle', figsize = (20,12), volume=True, mav=(7,14), style='classic')
```
